```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
rod_01 = pd.read_csv('rod_01.resu', header=4, delimiter= '\s+', index_col=False)
rod_02 = pd.read_csv('rod_02.resu', header=4, delimiter= '\s+', index_col=False)
rod_03 = pd.read_csv('rod_03.resu', header=4, delimiter= '\s+', index_col=False)
```

```python
data = {'D': np.concatenate((np.linspace(0,20,11), np.linspace(18,-20,20), np.linspace(-18,0,10)))}
result = pd.DataFrame(data)
result['LOG_LIN'] = -rod_01['RESULT_Z']/1000
result['GDEF_LOG'] = -rod_02['RESULT_Z']/1000
result['PETIT'] = -rod_03['RESULT_Z']/1000
# result['Analytical'] = 200000*(np.pi*(10**2))*np.log((100 + result['D'])/100)/1000
result['Linear'] = 200000*(np.pi*(10**2))/100*result['D']/1000
```

```python
ax = result.plot(x='D', y=[2,3])

plt.rcParams.update({'font.size': 12})
```
